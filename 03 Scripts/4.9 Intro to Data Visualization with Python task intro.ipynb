# 4.9 Intro to Data Visualization with Python

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Creating path
path = r'C:\Users\avadr\27-01-2020 Instacart Basket Analysis'
```

```python
# Importing data setfrom pickle format
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_.pkl'))
```

# Creating Bar Charts

```python
# Create bar chart
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Create bar chart and sorting by index
ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Create bar chart with differnet colours
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color = ['peachpuff'])
```

# Exporting Charts

```python
# Saving bar chart and to be able to call it any time by bar.figure in jupyter
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(edgecolor = 'black', color = ['peachpuff'])
plt.title('Busiest Day of the Week')
plt.xlabel('Day of the week')
plt.ylabel('Order Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Saving the bar chart outside the jupyter
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Checking if some product prices are higher the 100$
ords_prods_merge.loc[ords_prods_merge['prices'] > 100]
```

```python
# Turning values over 100 assigns missing value nan
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# Checking if all high values are replaced 
ords_prods_merge['prices'].max()
```

```python
ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# Saving into jupyter 
hist = ords_prods_merge['prices'].plot.hist(bins = 70, edgecolor = 'black', color =  ['peachpuff'])
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Exporting outside jupyter as png
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_orders_dow.png'))
```

# Creating Line Charts

## Sampling data

```python
# Creating a seed 
np.random.seed(4)
# Storing boolean values
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Storing 70% data into sample
big = ords_prods_merge[dev]
```

```python
# Storing 30% data into sample
small = ords_prods_merge[~dev]
```

```python
len(ords_prods_merge)
```

```python
len(big)+len(small)
```

```python
# Creating new df for spesific collums
df_3 = small[['order_hour_of_day','prices']]
```

```python
# Create Line Chart
prices_by_hour_line = sns.lineplot(data=df_3, x='order_hour_of_day', y='prices')
# Define title and labels
plt.title('Prices By hour')
plt.ylabel('Prices')
plt.xlabel('Time of Day')
plt.show()
```

```python
prices_by_hour_line .figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'prices_by_hour_line .png'))
```

```python
# Creating new df for spesific collums
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Creating line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
plt.title('Prices By day')
plt.ylabel('Prices')
plt.xlabel('Orders')
plt.show()
```

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_dow.png'))
```
