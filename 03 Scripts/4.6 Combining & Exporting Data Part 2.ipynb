#### By product_id and as a full merge, the data can be merged most effectively.

```python
#Drop unnamed columns in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
#Drop unnamed columns in df_ords_prod
df_ords_prod = df_ords_prod.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
df_prods
```

```python
# Merge the two dataframes
df_merged = df_prods.merge(df_ords_prod, on = 'product_id', indicator = True)
```
