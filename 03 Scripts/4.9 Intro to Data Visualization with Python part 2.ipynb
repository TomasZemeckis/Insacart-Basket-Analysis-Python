```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Creating path
path = r'C:\Users\avadr\27-01-2020 Instacart Basket Analysis'
```

```python
# Importing data setfrom pickle format
final = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'final_.pkl'))
```

```python
final.info()
```

```python
# Creating histogram
final['order_hour_of_day'].plot.hist(bins = 24)
```

#### The busiest period of day for the order is about 9 a.m. to 4 p.m.

```python
# Saving into jupyter 
hist = final['order_hour_of_day'].plot.hist(bins = 24, edgecolor = 'black', color = ['peachpuff'])
plt.title("Order Frequency of the Hour")
plt.xlabel("Hour of the Day ")
plt.ylabel("Order Frequency")
plt.show()
```

```python
# Exporting outside jupyter as png
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_orders_hour_of_day.png'))
```

```python
# Create bar chart
final['loyalty_flag'].value_counts().plot.bar()
```

```python
# Saving into jupyter 
bar = final['loyalty_flag'].value_counts().plot.bar(edgecolor = 'black', color = ['peachpuff'])
plt.title("Customer Loyalty")
plt.ylabel("Number of Customer")
plt.xticks(rotation=0)
plt.show()
```

```python
# Exporting outside jupyter as png
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty_flag.png'))
```

```python
# Creating a seed 
np.random.seed(4)
# Storing boolean values
dev = np.random.rand(len(final)) <= 0.7
```

```python
dev
```

```python
# Storing 70% data into sample
big = final[dev]
```

```python
# Storing 30% data into sample
small = final[~dev]
```

```python
len(final)
```

```python
len(big) + len(small)
```

```python
# Creating new df for spesific collums
df = small[['order_hour_of_day','prices']]
```

```python
# Creating line chart
line = sns.lineplot(data = df, x = 'order_hour_of_day',y = 'prices',)
```

```python
# Saving as PNG
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_prices_age.png'))
```

```python
# Creating new df for spesific collums
df_2 = small[['Age','dependants']]
```

```python
# Creating line chart
line_2 = sns.lineplot(data = df_2, x = 'Age',y = 'dependants',)
```

```python
# Saving as PNG
line_2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_age_dependents.png'))
```

#### It appears that dependant number is not predicted by age.

```python
# Saving into Jupyter
scatt = sns.scatterplot(x = 'Age', y = 'income',data = final, color = ['peachpuff'])
```

```python
# Saving as PNG
scatt.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_age_income.png'))
```
